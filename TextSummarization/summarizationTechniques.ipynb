### Prompt Template to summarize

```python
from langchain.prompts import ChatPromptTemplate
from langchain.chains import LLMChain
from langchain_core.output_parsers import StrOutputParser
```

```python
parser = StrOutputParser()
```

```python
generic_prompt = '''Write a summary of the following speech :{speech} and translate into given language :{language}'''
prompt = ChatPromptTemplate.from_messages(
    [
        ("system",generic_prompt),
        ("user","{speech},{language}")
    ]
)
```

```python
llmchain = LLMChain(llm=llm,prompt=prompt)
```

```python
llmchain.run({"speech":speech,"language":"Hindi"})
```

```python
second_chain = prompt|llm|parser
```

```python
second_chain.invoke({"speech":speech,"language":"Hindi"})
```
